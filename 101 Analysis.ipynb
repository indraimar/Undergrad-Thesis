```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Calculation of Total US Aid to Contras from FY1982 through FY1990

```python
d = {'Fiscal Year': [1982, 1983, 1984, 1985, 1986, 1987, 1988, 1989, 1990], 
     'Nonlethal Aid (Millions of US$)': [0, 0, 0, 27, 13, 30, 32.9, 27, 49.75], 
     'Lethal Aid (Millions of US$)':[19.0, 29, 24, 0, 0, 70, 0, 0, 0], 
     'Total Aid (Millions of US$)': [19, 29, 24, 27, 13, 100, 32.9, 27, 49.75]}
us_aid_df = pd.DataFrame(d)
us_aid_df_1 = us_aid_df.drop(columns = ['Total Aid (Millions of US$)'])
```

```python
us_aid_df_total = us_aid_df.append({'Fiscal Year' : 'Total' , 'Nonlethal Aid (Millions of US$)' : us_aid_df['Nonlethal Aid (Millions of US$)'].sum(), 'Lethal Aid (Millions of US$)' : us_aid_df['Lethal Aid (Millions of US$)'].sum(), 'Total Aid (Millions of US$)' : us_aid_df['Total Aid (Millions of US$)'].sum()} , ignore_index=True)
us_aid_df_total
#us_aid_df_total.to_csv('US AID DF TOTAL.csv', index = False)
```

```python
us_aid_df_total_2 = us_aid_df.T
us_aid_df_total_2.columns = us_aid_df_total_2.iloc[0]
us_aid_df_total_2 = us_aid_df_total_2.drop(us_aid_df_total_2.index[0])
us_aid_df_total_2.columns = us_aid_df_total_2.columns.astype(int)
us_aid_df_total_2
```

```python
us_aid_df_2 = us_aid_df_1.melt('Fiscal Year', var_name='Aid Type',  value_name='Amount (Millions of US$)')
plt.figure(figsize=(16, 8))
us_both_aid_barplot = sns.barplot(x="Fiscal Year", y="Amount (Millions of US$)", hue='Aid Type', data=us_aid_df_2, palette = 'rocket')
for p in us_both_aid_barplot.patches:
    us_both_aid_barplot.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
```

```python
g = sns.catplot(x="Fiscal Year", y="Amount (Millions of US$)", hue='Aid Type', data=us_aid_df_2,
                height=10, kind="bar", palette="muted")
```

```python
plt.figure(figsize=(16, 8))
us_both_aid_factorplot = sns.factorplot(x="Fiscal Year", y="Amount (Millions of US$)", hue='Aid Type', data=us_aid_df_2, palette = 'rocket')
#us_both_aid_factorplot.savefig("US BOTH AID FACTORPLOT.png")
```

```python
plt.figure(figsize=(16, 8))
us_total_aid_factorplot = sns.factorplot(x="Fiscal Year", y="Total Aid (Millions of US$)", data=us_aid_df, color = 'g')
#us_total_aid_factorplot.savefig("US TOTAL AID FACTORPLOT.png")
```

## Economic Cost of Contra War FY1981-1985

```python
z = {'Fiscal Year': [1981, 1982, 1983, 1984, 1985], 
     'Material Damage (Millions of US$)': [0.5, 3.9, 10.8, 57.5, 24.4], 
     'Production Losses (Millions of US$)':[0.9, 3.5, 21.1, 97.4, 159.7], 
     'Total Damage/Losses (Millions of US$)': [1.4, 7.4, 31.9, 154.9, 184.1]}
econ_cost_war_df = pd.DataFrame(z)
econ_cost_war_df = econ_cost_war_df.append({'Fiscal Year' : 'Total' , 'Material Damage (Millions of US$)' : econ_cost_war_df['Material Damage (Millions of US$)'].sum(), 'Production Losses (Millions of US$)' : econ_cost_war_df['Production Losses (Millions of US$)'].sum(), 'Total Damage/Losses (Millions of US$)' : econ_cost_war_df['Total Damage/Losses (Millions of US$)'].sum()} , ignore_index=True)
econ_cost_war_df
#econ_cost_war_df.to_csv('ECON COST WAR DF.csv', index = False)
```

```python
x = {'Fiscal Year': [1981, 1982, 1983, 1984, 1985],
     'Credit Blocks (Millions of US$)': [8.2, 38.3, 61.3, 92.1, 73.0], 
     'Trade Embargo (Millions of US$)':[0, 0, 11.7, 15.0, 50.0],
    'Total Losses (Millions of US$)' : [8.2, 38.3, 73, 107.1, 123]}
econ_cost_pres_df = pd.DataFrame(x)
econ_cost_pres_df = econ_cost_pres_df.append({'Fiscal Year' : 'Total' , 'Credit Blocks (Millions of US$)' : econ_cost_pres_df['Credit Blocks (Millions of US$)'].sum(), 'Trade Embargo (Millions of US$)' : econ_cost_pres_df['Trade Embargo (Millions of US$)'].sum(), 'Total Losses (Millions of US$)' : econ_cost_pres_df['Total Losses (Millions of US$)'].sum()} , ignore_index=True)
econ_cost_pres_df
#econ_cost_pres_df.to_csv('ECON COST PRES DF.csv', index = False)
```

## Casualties

```python
y = {'Year': [1981, 1982, 1983, 1984, 1985],
     'Deaths': [53, 114, 1030, 1339, 1463], 
     'Wounded':[13, 52, 1323, 1302, 1852],
    'Missing/Kidnapped' : [7, 91, 1153, 1085, 1455]}
civilians_df = pd.DataFrame(y)
civilians_df = civilians_df.append({'Year' : 'Total' , 'Deaths' : civilians_df['Deaths'].sum(), 'Wounded' : civilians_df['Wounded'].sum(), 'Missing/Kidnapped' : civilians_df['Missing/Kidnapped'].sum()} , ignore_index=True)
civilians_df
#civilians_df.to_csv('CIVILIAN TOLL DF.csv', index = False)
```
